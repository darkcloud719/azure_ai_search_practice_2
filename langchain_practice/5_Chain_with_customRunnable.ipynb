```python
# TASK - 1 [prompt]

prompt_template = ChatPromptTemplate.from_messages([
    ("system","You are a helpful assistant"),
    ("human","{input}")
])

# TASK - 2 [LLM]

# llm_openai

# TASK - 3 [str Parser]

str_parser = StrOutputParser()

# TASK - 4 [Custom Runnable]

def dictionary_maker(text:str) -> dict:

    return {"text": text}

dictionary_maker_runnable = RunnableLambda(dictionary_maker)

# TASK - 5 [Template For Post]

prompt_post = ChatPromptTemplate.from_messages(
    messages=[
        ("system","You are a social media post generator."),
        ("human","Create a post for the following text for LinkedIn: {text}")
    ]
)

# TASK - 6 [Generate the Post]

# TASK - 7 [Str Parser]

str_parser = StrOutputParser()

# CHAIN

chain = prompt_template | llm_openai | str_parser | dictionary_maker_runnable | prompt_post | llm_openai | str_parser

chain.invoke("What is the capital of France?")


```

# Langchain Practice - Translate the following Mandarin to English

```python
from langchain_openai import AzureChatOpenAI
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from dotenv import load_dotenv
from rich import print
from rich.logging import RichHandler
from rich.table import Table
from rich.console import Console
import time
import os
import logging
import warnings

warnings.filterwarnings("ignore")

logger = logging.getLogger()
logger.setLevel(logging.INFO)

console_handler = RichHandler()
console_handler.setLevel(logging.INFO)

file_handler = logging.FileHandler("app.log", mode="a", encoding="utf-8")
file_handler.setLevel(logging.INFO)

formatter = logging.Formatter(
    "%(asctime)s - %(name)s - %(levelname)s - %(message)s",
    datefmt="%Y-%m-%d %H:%M:%S"
) 

file_handler.setFormatter(formatter)

logger.addHandler(console_handler)
logger.addHandler(file_handler)

console = Console()

load_dotenv()

AZURE_OPENAI_API_KEY = os.getenv("AZURE_OPENAI_API_KEY")
AZURE_OPENAI_API_VERSION = os.getenv("AZURE_OPENAI_API_VERSION")
AZURE_OPENAI_API_ENDPOINT = os.getenv("AZURE_OPENAI_API_ENDPOINT")
AZURE_OPENAI_API_DEPLOYMENT_4O = os.getenv("AZURE_OPENAI_API_DEPLOYMENT_4O")

llm_openai = AzureChatOpenAI(
    azure_endpoint=AZURE_OPENAI_API_ENDPOINT,
    deployment_name=AZURE_OPENAI_API_DEPLOYMENT_4O,
    openai_api_version=AZURE_OPENAI_API_VERSION,
    openai_api_key=AZURE_OPENAI_API_KEY
)

# Prompt

prompt = ChatPromptTemplate.from_messages([
    ("system","You are a professional translator. Translate Mandarin to English."),
    ("human","{input}")
])

# Parser 
parser = StrOutputParser()

# Chain 
chain = prompt | llm_openai | parser

# Invoke

msg = "教授跟我說:如果學東西要別人教你，你才開始學，那你就太弱了"
result = chain.invoke({"input": msg})

table = Table(title="Translated Text")
table.add_column("Input", style="cyan")
table.add_column("Output", style="magenta")
table.add_row(msg, result)
console.print(table)


```

```python
from langchain_openai import AzureChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnableLambda
from dotenv import load_dotenv
import os
import warnings

warnings.filterwarnings("ignore")

logger = logging.getLogger()
logger.setLevel(logging.INFO)

console_handler = RichHandler()
console_handler.setLevel(logging.INFO)

file_handler = logging.FileHandler("app.log", mode="a", encoding="utf-8")
file_handler.setLevel(logging.INFO)

formatter = logging.Formatter(
    "%(asctime)s - %(name)s - %(levelname)s - %(message)s",
    datefmt="%Y-%m-%d %H:%M:%S"
)

file_handler.setFormatter(formatter)

logger.addHandler(console_handler)
logger.addHandler(file_handler)

console = Console()

load_dotenv()

AZURE_OPENAI_API_KEY = os.getenv("AZURE_OPENAI_API_KEY")
AZURE_OPENAI_API_ENDPOINT= os.getenv("AZURE_OPENAI_API_ENDPOINT")
AZURE_OPENAI_API_DEPLOYMENT_4O = os.getenv("AZURE_OPENAI_API_DEPLOYMENT_4O")
AZURE_OPENAI_API_VERSION = os.getenv("AZURE_OPENAI_API_VERSION")

llm_openai = AzureChatOpenAI(
    azure_endpoint=AZURE_OPENAI_API_ENDPOINT,
    deployment_name=AZURE_OPENAI_API_DEPLOYMENT_4O,
    openai_api_key=AZURE_OPENAI_API_KEY,
    openai_api_version=AZURE_OPENAI_API_VERSION
)

parser = StrOutputParser()

prompt_explain = ChatPromptTemplate.from_messages([
    ("system","You are a senior software engineer. Explain the problem clearly."),
    ("human","{input}")
])

prompt_summary = ChatPromptTemplate.from_messages([
    ("system","Give practical debugging advice."),
    ("human","{text}")
])

prompt_advice = ChatPromptTemplate.from_messages([
    ("system","Give practical debugging advice."),
    ("human","{text}")
])

prompt_teacher = ChatPromptTemplate.from_messages([
    ("system","Explain like a mentor teaching a junior developer."),
    ("human","{text}")
])

def to_dict(text:str) -> dict:
    return {"text": text}

to_dict_runnable = RunnableLambda(to_dict)


chain = (
    prompt_explain
    | llm_openai
    | parser
    | to_dict_runnable
    | prompt_summary
    | llm_openai
    | parser
    | to_dict_runnable
    | prompt_advice
    | llm_openai
    | parser
    | to_dict_runnable
    | prompt_teacher
    | llm_openai
    | parser
)

query = "LangChain 為什麼會出現key error?"
result = chain.invoke({"input": query})

logger.info(f"Final Result: {result}")



```

# Langchain Practice 3 - Reservation

```python
from langchain_openai import AzureChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.runnables import RunnableLambda
from pydantic import BaseModel, Field
from typing import Optional
from datetime import datetime, timedelta
from dotenv import load_dotenv
import os

load_dotenv()

logger = logging.getLogger()
logger.setLevel(logging.INFO)

console_handler = RichHandler()
console_handler.setLevel(logging.INFO)

file_handler = logging.FileHandler("app.log", mode="a", encoding="utf-8")
file_handler.setLevel(logging.INFO)

formatter = logging.Formatter(
    "%(asctime)s - %(name)s - %(levelname)s - %(message)s",
    datefmt="%Y-%m-%d %H:%M:%S"
)

file_handler.setFormatter(formatter)

logger.addHandler(console_handler)
logger.addHandler(file_handler)

console = Console()

class Reservation(BaseModel):
    restaurant: str = Field(description="The name of the restaurant")
    date: str = Field(description="The date of the reservation in YYYY-MM-DD format")
    time: str = Field(description="The time of the reservation in HH:MM format")
    people: int = Field(description="The number of people for the reservation")
    special_request: Optional[str] = Field(default=None, description="Any special requests for the reservation")

load_dotenv()

AZURE_OPENAI_API_KEY = os.getenv("AZURE_OPENAI_API_KEY")
AZURE_OPENAI_API_VERSION = os.getenv("AZURE_OPENAI_API_VERSION")
AZURE_OPENAI_API_ENDPOINT = os.getenv("AZURE_OPENAI_API_ENDPOINT")
AZURE_OPENAI_API_DEPLOYMENT_4O = os.getenv("AZURE_OPENAI_API_DEPLOYMENT_4O")

llm_openai = AzureChatOpenAI(
    azure_endpoint=AZURE_OPENAI_API_ENDPOINT,
    deployment_name=AZURE_OPENAI_API_DEPLOYMENT_4O,
    openai_api_key=AZURE_OPENAI_API_KEY,
    openai_api_version=AZURE_OPENAI_API_VERSION
)

prompt_parse = ChatPromptTemplate.from_messages([
    ("system","You are a restaurant booking assistant. Extract all relevant information from user request."),
    ("human","{input}")
])

structured_llm = llm_openai.with_structured_output(Reservation)

def check_availability(reservation: Reservation) -> Reservation:
    # Simulate checking availability (for demonstration purposes)
    full_times = ["19:00","20:00"]
    if reservation.time in full_times:
        hour, minute = map(int, reservation.time.split(":"))
        new_hour = (hour + 1) % 24
        reservation.time = f"{new_hour:02d}:{minute:02d}"
        if reservation.special_request:
            reservation.special_request += " (Note: Original time was fully booked, new time is one hour later.)"
        else:
            reservation.special_request = "Note: Original time was fully booked, new time is one hour later."

    return reservation


check_availability_runnable = RunnableLambda(check_availability)

prompt_confirm = ChatPromptTemplate.from_messages([
    ("system","You are a friendly booking assistant. Generate a confirmation message based on the reservation details."),
    ("human","Reservation details: {text}")
])

str_parser = StrOutputParser()

chain = (
    prompt_parse
    | structured_llm
    | check_availability_runnable
    | RunnableLambda(lambda r: {"text": str(r.dict())})
    | prompt_confirm
    | llm_openai
    | str_parser
)


# user_input = "I want to book a table for 4 people at The Italian Place on 2024-07-15 at 19:00. I would like a quiet corner if possible."
user_input = "I want to book a table for 4 people at The Italian Place"
result = chain.invoke({"input":user_input})

print("=== User Input ===")
print(user_input)
print("\n=== Reservation Confirmation ===")
print(result)

```
